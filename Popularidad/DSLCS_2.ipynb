```python
%pip install geopy
%pip install geopandas
import geopy
import geopandas
from geopandas.tools import geocode
from geopy.geocoders import Nominatim

# Set the custom user agent
user_agent = "MyAwesomeApp/1.0 (gbekoilias@gmail.com)"
geolocator = Nominatim(user_agent=user_agent)

map_data = df1
map_data.dropna(axis=0, inplace=True)

# Geocode using the custom geolocator
map_data['geometry'] = map_data['País'].apply(lambda x: geolocator.geocode(x) if x else None)
map_data = map_data[map_data['geometry'].notnull()]

# Extract Latitude and Longitude
map_data['Latitude'] = map_data['geometry'].apply(lambda loc: loc.latitude if loc else None)
map_data['Longitude'] = map_data['geometry'].apply(lambda loc: loc.longitude if loc else None)
```

```python
map_data.head()
```

```python
m3 = folium.Map(location=[39.326234, -4.838065], tiles='OpenStreetMap', zoom_start=3)

def color_producer(val):
    if val <= 50:
        return 'red'
    else:
        return 'green'

for i in range(0, len(map_data)):
    folium.Circle(
        location=[map_data.iloc[i]['Latitude'], map_data.iloc[i]['Longitude']],
        radius=5000 * map_data.iloc[i]['data science'],
        color=color_producer(map_data.iloc[i]['data science'])
    ).add_to(m3)

m3.save('map3.html')
```
